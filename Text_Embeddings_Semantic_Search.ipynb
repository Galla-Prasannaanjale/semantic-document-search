```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
documents = [
    "Machine learning is used for data analysis",
    "Deep learning is a subset of machine learning",
    "Cybersecurity focuses on protecting networks",
    "Natural language processing deals with text data",
    "Artificial intelligence enables intelligent systems"
]
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
doc_embeddings = model.encode(documents)
```

```python
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(doc_embeddings)
```

```python
def search(query):
    query_embedding = model.encode([query])
    distances, indices = index.search(query_embedding, k=1)
    return documents[indices[0][0]]
```

```python
query = "techniques for working with text"
print("Result:", search(query))
```
